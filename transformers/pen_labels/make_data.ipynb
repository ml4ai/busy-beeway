```python
import numpy as np

import h5py
```

```python
with h5py.File("AdroitHandPen-v1_pref_b.hdf5") as f:
    rng = np.random.default_rng(8675309)
    r_idx = rng.permutation(f["actions"].shape[0])[:200]

    acts = f["actions"][:][r_idx, ...]
    acts_2 = f["actions_2"][:][r_idx, ...]
    am = f["attn_mask"][:][r_idx, ...]
    am_2 = f["attn_mask_2"][:][r_idx, ...]
    rwds = f["rewards"][:][r_idx, ...]
    rwds_2 = f["rewards_2"][:][r_idx, ...]
    sts = f["states"][:][r_idx, ...]
    sts_2 = f["states_2"][:][r_idx, ...]
    ts = f["timesteps"][:][r_idx, ...]
    ts_2 = f["timesteps_2"][:][r_idx, ...]
    lab = f["labels"][:][r_idx, ...]

    with h5py.File("AdroitHandPen-v1_pref_small.hdf5", "a") as g:
        g.create_dataset("actions", data=acts)
        g.create_dataset("actions_2", data=acts_2)
        g.create_dataset("attn_mask", data=am)
        g.create_dataset("attn_mask_2", data=am_2)
        g.create_dataset("rewards", data=rwds)
        g.create_dataset("rewards_2", data=rwds_2)
        g.create_dataset("states", data=sts)
        g.create_dataset("states_2", data=sts_2)
        g.create_dataset("timesteps", data=ts)
        g.create_dataset("timesteps_2", data=ts_2)
        g.create_dataset("labels", data=lab)
```
